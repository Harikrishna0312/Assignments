```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/harik/Downloads/sales_data_with_discounts (1).csv")
numerical_cols = df.select_dtypes(include=['int64', 'float64'])
numerical_cols.describe()
```

#### 1.Histogram

```python
numerical_cols.hist(figsize=(10,8))
plt.show()
```

##### Analysis

Histograms reveal whether data is normally distributed, left-skewed, or right-skewed.

Sales values often show right skewness, indicating a few high-value transactions.

Extreme bars may indicate the presence of outliers.

#### 2.Box plot

```python
numerical_cols.boxplot(figsize=(10,6))
plt.show()
```

##### Analysis

Box plots clearly identify outliers beyond the whiskers.

A wide interquartile range (IQR) suggests high variability in sales or discounts.

Extreme outliers may require further preprocessing such as capping or removal.

##### 3.Bar Chart Analysis for Categorical Columns

```python
categorical_cols = df.select_dtypes(include=['object'])

for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```

##### Analysis

Bar charts show the frequency distribution of categories such as Region, Product Category, or Customer Segment.

Dominant categories indicate higher sales concentration.

Uneven distribution suggests possible business dependency on specific segments

##### Data Preprocessing Insights

Missing values can be handled using mean/median imputation for numerical data.

Outliers identified through box plots can be treated using IQR method or winsorization.

Categorical variables can later be encoded using Label Encoding or One-Hot Encoding.
